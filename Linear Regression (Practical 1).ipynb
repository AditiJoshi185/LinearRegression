```python
data_no_multicollinearlity = data_cleaned.drop(['Year'],axis=1)
data_no_multicollinearlity
```

## Create Dummies


```python
data_with_dummies = pd.get_dummies(data_no_multicollinearlity, drop_first = True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['Mileage', 'EngineV', 'log_price', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

# Linear Regression Model

```python
target = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

## Split Data into Training and Testng

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_scaled,target, test_size=0.2,random_state=365)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets y_train',size=18)
plt.ylabel('Predictions y_hat',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF',size=18)
```

```python
reg.score(x_train,y_train)
```

## Finding Weights and Bias

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets y_test',size=18)
plt.ylabel('Predictions y_hat_test',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(y_hat_test,columns = ['Predictions']) # dataframe performance
df_pf.head()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test),columns = ['Predictions']) # dataframe performance
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop='True')
```
